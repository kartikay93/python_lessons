Simple calculator using if-else statement

```python
num1 = float(input("Enter the first number"))
num2 = float(input("Enter the second number"))

print("Select the operation to perform: +, -, *, /")
operation = input("enter the operation:")
if operation == '+':
    print(num1, "+", num2, "=", num1+num2)
elif operation == '-':
    print(num1, "-", num2, "=", num1-num2)
elif operation == '*':
    print(num1, '*', num2, "=", num1*num2)
elif operation == '/':
    if num2 != 0:
        print(num1, '/', num2, "=", num1/num2)
    else:
        print("Error: Division by zero is not allowed")
else:
    print("Invalid operation")
```

## Printing Prime number from 1 to n


```python
num = int(input("Enter the number to which you want to find the prime numbers: "))

if num > 1:
    for current_num in range(2, num+1):
        for i in range(2, current_num):
            if current_num % i == 0:
                break
        else:
            print(current_num)
```

### SETS

```python
#Set A , Set B
A = {1,2,3,4,5,}
B = {4,5,6,7,8,}
union = A.union(B)
print(union)
```

```python
intersection = A.intersection(B)
print(intersection)
```

```python
A.intersection_update(B)
print(A)
B.intersection_update(A)
print(B)
```

```python
#Set A , Set B
A = {1,2,3,4,5,}
B = {4,5,6,7,8,}
A.difference(B)
print(A)
B.difference(A)
print(B)
```
